```python
  def get_caption(line:str):
    line = line.split("|")

    file_dir = line[0]
    audio_file_name = file_dir.split("/")[-1]
    original_script = line[1]
    normalized_script = line[2]
    audio_duration = line[3]

    return file_dir, audio_file_name, original_script, normalized_script, audio_duration
```

```python
utterances = {}
```

```python
# Open the file in read mode
with open("/content/drive/MyDrive/SP Cup 2024/transcript.txt", 'r') as file:
    # Read all lines into a list
    lines = file.readlines()

# Process each line as needed
for line in lines:
    # print(line.strip())  # Example: Print each line after removing leading and trailing whitespaces
    audio_meta = get_caption(line)

    utterances[audio_meta[1]] = audio_meta[2]


```

```python
from gtts import gTTS
```

```python
for k,v in enumerate(utterances):
  print(v,utterances[v] )
  if k == 30:
    break
```

```python
save_path = "/content/drive/MyDrive/SP Cup 2024/Generated_Audio/"
```

```python
for j,v in enumerate(utterances):

  if j < 100:
    i = 1
    file_name = f"spk_{i}-{j}"
    # generate audio
    tts = gTTS(utterances[v], lang='fr') # French
    tts.save(f'{save_path}{file_name}.mp3')

  elif j >= 100 and j < 200:
    i = 2
    file_name = f"spk_{i}-{j}"
    # generate audio
    tts = gTTS(utterances[v], lang='fr',tld="ca") # French in Canada
    tts.save(f'{save_path}{file_name}.mp3')

  elif j == 200:
    break
```

```python
import pyttsx3

# Initialize the TTS engine
engine = pyttsx3.init()

# Set properties, including the male voice
engine.setProperty('gender', 'male')

# Set the text you want to convert to speech
text = "Hello, this is an example of using a male voice in TTS."

# Save the speech to an audio file
engine.save_to_file(text, 'output.wav')

# Run the TTS engine to convert text to speech
engine.runAndWait()
```
